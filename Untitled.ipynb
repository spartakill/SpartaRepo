```python
import pandas as pd
import numpy as np
```

```python
nme = ['Tommy', 'Ruth', 'Ruth', 'Taslima', 'Isak', 'Killian', 'Yuvraj']
deg = ["Biochemistry", "English", "English", "English", "ComputerScience", None, "Biology"]
age = [24, 28, None, 25, 30, 25, 23]

# dictionary of lists
dict = {'name': nme, 'degree': deg, 'age': age}
df = pd.DataFrame(dict)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
## filling missing values with a value
df2 = df.fillna(value = 0)
df2
```

```python
## filling null value with previous value row wise
df3 = df.fillna(method = 'pad')
df3
```

```python
## filling null value with previous value row wise
df4 = df.ffill()
df4
```

```python
## filling null values with coming value row wise
df5 = df.bfill()
df5
```

```python
## filling null values with the previous value column wise
df6 = df.ffill(axis = 1)
df6
```

```python
## filling null values with coming value column wise
df7 = df.bfill(axis = 1)
df7
```

```python
## filling different values in null in different columns
df8 = df.fillna({'degree': 'Biomed'})
df8
```

```python
## filling null value with the mean, max and min of a column
df9 = df8.fillna(value = df['age'].mean())
df9
```

```python
## filling null value with the mean, max and min of a column
df10 = df8.fillna(value = df['age'].max())
df10
```

```python
## filling null value with the mean, max and min of a column
df11 = df8.fillna(value = df['age'].min())
df11
```

```python
## replacing values in a dataframe with values
df12 = df.replace(to_replace = np.nan, value = 5.0)
df12
```

```python
## using interpolate
df['age'] = df['age'].interpolate()
df
```
