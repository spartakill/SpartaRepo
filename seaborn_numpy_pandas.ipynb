```python
flights = sns.load_dataset('flights')
flights
```

```python
f_piv = flights.pivot(index = 'year', columns = 'month', values = 'passengers')
f_piv
```

```python
f_stack = f_piv.stack()
f_stack # changes it to a mutlilevel index where year and month are the index
```
