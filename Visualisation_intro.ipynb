```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
import warnings # removes red boxed future warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
plt.title("FRMI signal across timepoint")
```

```python
plt.plot(fmri.timepoint, fmri.signal)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.barplot(data=tips, x='day', y='total_bill', hue='time', errorbar=None)
plt.title("AVG tips based on each day of the week")
```

```python
plt.bar(tips.day, tips.total_bill) # prints all of the values on top of eachother so only the tallest bar is shown
gtips = tips.groupby('day')
gtips.describe()
```

```python
iris = sns.load_dataset("iris")
g = sns.jointplot(data=iris, x='sepal_width', y='sepal_length')
g.plot(sns.scatterplot, sns.histplot)
```
